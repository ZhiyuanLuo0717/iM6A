```python
import glob,os
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
path = "Outputs"
```

```python
file = glob.glob(os.path.join(path, "*.csv"))
```

```python
for f in file:
    print(f[8:11])
```

```python
Name = []
Intercept = []
Slope = []
R2 = []
Pvalue = []

for f in file:
    df = pd.read_csv(f)    
    name = f[8:11]
    x = np.asarray(df["Score"], dtype=float)
    y = np.asarray(df["Value"], dtype=float)
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    
    r2 = model.rsquared    
    intercept = model.params[0]
    slope = model.params[1]    
    pvalue = model.pvalues[0]
    
    Name.append(name)
    Slope.append(slope)
    Intercept.append(intercept)
    R2.append(r2)
    Pvalue.append(pvalue)
```

```python
List = [Name, Slope, Intercept, Pvalue, R2]
df = pd.DataFrame(List)
df = df.T
df.rename(columns={0:"Name", 1:"Slope", 2:"Intercept", 3:"Pvalue", 4:"R2"}, inplace = True)
```

```python
df
```

```python
df = df.sort_values(by=["Slope"], axis=0, ascending=[False]) 
df = df.reset_index(drop = True)
```

```python
df.to_csv("Mouse_Codon_Regression.csv", index=0)
```
