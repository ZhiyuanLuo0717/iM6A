## Grab sequence

```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority.txt", sep="\t", header=None)
```

```python
Data.drop([0,11,13,14,15], axis=1, inplace=True)
```

```python
Data
```

```python
Data[1] = Data[12]
```

```python
Data.drop([12], axis=1, inplace=True)
```

```python
Data
```

```python
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
```

```python
Data
```

### Select coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data
```

### Keep standard chromosomes

```python
Data = Data[:22357]
```

### exonCount > 1

```python
Data = Data[Data.exonCount > 1]
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["LastExonStart"] = 0
Data["LastExonEnd"] = 0
```

```python
Data
```

```python
for i in range(len(Data)):
    if Data.loc[i, "strand"] == "+":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[-2]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[-2]
    elif Data.loc[i, "strand"] == "-":
        Data.loc[i,("LastExonStart")] = Data.loc[i, "exonStarts"].split(",")[0]
        Data.loc[i,("LastExonEnd")] = Data.loc[i, "exonEnds"].split(",")[0]
```

```python
Data
```

```python
Data.drop(["exonCount", "exonStarts", "exonEnds"], axis=1, inplace=True)
```

```python
Data
```

```python
Data["LastExonStart"] = Data["LastExonStart"].astype(int)
Data["LastExonEnd"] = Data["LastExonEnd"].astype(int)
```

```python
StopCodon = pd.read_csv("StopCodon.csv")
```

```python
StopCodon
```

```python
Data = pd.merge(Data, StopCodon, how="inner", on=["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "LastExonStart", "LastExonEnd"])
```

```python
Data
```

```python
Group = pd.read_csv("Group.csv")
```

```python
Group
```

```python
Data = pd.merge(Data, Group, how="inner", on=["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "LastExonStart", "LastExonEnd"])
```

```python
Data
```

```python
Data = Data[Data["Group"]==2]
Data.drop(["StopCodon","exonCount","Group"], axis=1, inplace=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Bed = Data[["chrom","LastExonStart","LastExonEnd"]]
```

```python
Bed
```

```python
Bed.to_csv("mm10_LastExon.bed", sep="\t", header=False, index=False)
```
