### Fasta2CSV

```python
from Bio import SeqIO
import numpy as np
import pandas as pd
```

```python
meta=[]
sequence=[]
```

```python
seq = ('mm10.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.read_csv("mm10.csv")
```

```python
Data
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result
```

```python
Result["Length"] = Result["txEnd"] - Result["txStart"]
```

```python
Result
```

```python
Data = Result
```

```python
Data 
```

```python
Data = Data.sort_values(by=["chrom","txStart"], axis=0, ascending=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.to_csv("mm10_Fasta.csv", index=0)
```
