```python
import glob,os
import numpy as np
import pandas as pd
```

### Read multiple files

```python
path = "Outputs/m3cRRACH"
```

```python
file = glob.glob(os.path.join(path, "*.bed"))
```

```python
d = []
for f in file:
    d.append(pd.read_table(f))
```

### Delete empty dataframe

```python
for i in range(len(d))[::-1]:
    if len(d[i]) == 0:
        del d[i]
```

```python
Data = pd.concat(d)
```

```python
Data = Data.sort_values(by=["chrom","Start"], axis=0, ascending=True)
Data = Data.reset_index(drop = True)
```

```python
Data
```

### Get iM6A sites

#### For positive strand

```python
idx = (Data["strand"] == "+")
```

```python
Positive = Data.loc[idx, ["name", "chrom", "strand", "Start", "End", "Probabilty"]]
```

```python
Positive["Start"] = Positive["Start"] + 1
Positive["End"] = Positive["End"] + 1
```

```python
Positive
```

#### For negative strand

```python
idx = (Data["strand"] == "-")
```

```python
Negative = Data.loc[idx, ["name", "chrom", "strand", "Start", "End", "Probabilty"]]
```

```python
Negative["Start"] = Negative["Start"] - 1
Negative["End"] = Negative["End"] - 1
```

```python
Negative
```

```python
iM6A = pd.concat([Positive, Negative], axis=0)
```

```python
iM6A = iM6A.sort_values(by=["chrom","Start"], axis=0, ascending=True)
iM6A = iM6A.reset_index(drop = True)
```

```python
iM6A
```

```python
iM6A = iM6A[["chrom", "Start", "End", "name", "Probabilty", "strand"]]
```

```python
iM6A["End"] = iM6A["End"] + 1
```

```python
iM6A
```

### Get RAC sites for iM6A

```python
RAC = pd.read_table("RRACH.bed", header=None)
```

```python
RAC.columns = ["chrom", "Start", "End", "name", "Score", "strand"]
```

```python
RAC
```

```python
RAC.drop(["name", "Score"], axis=1, inplace=True)
```

```python
RAC = pd.merge(RAC, iM6A, how="inner", on=["chrom", "Start", "End", "strand"])
```

```python
RAC
```

```python
RAC = RAC[["chrom", "Start", "End", "name", "Probabilty", "strand"]]
```

```python
RAC = RAC.drop_duplicates()
```

```python
RAC = RAC.reset_index(drop = True)
```

```python
RAC
```

```python
iM6A = RAC
```

### Label predicted RAC sites

#### all3combined m6A-CLIP

```python
CLIP = pd.read_table("Annotation/RRACH_precise_m6A_sites_all3combined_mm10.bed", header=None)
CLIP.columns = ["chrom", "Start", "End", "name", "Score", "strand"]
CLIP = CLIP.sort_values(by=["chrom","Start"], axis=0, ascending=True)
CLIP = CLIP.reset_index(drop = True)
```

```python
Overlap = pd.merge(iM6A, CLIP, how="inner", on=["chrom", "Start", "End", "strand"])
Overlap.drop(["name_y"], axis=1, inplace=True)
Overlap.rename(columns={"name_x":"name"}, inplace = True)
```

```python
Overlap
```

```python
iM6A = pd.merge(iM6A, Overlap, how="left", on=["chrom", "Start", "End", "name", "Probabilty", "strand"])
iM6A.rename(columns={"Score":"3c"}, inplace = True)
iM6A = iM6A.fillna(0)
iM6A["3c"] = iM6A["3c"].astype(int)
```

```python
iM6A
```

```python
iM6A.to_csv("iM6A_m3cRRACH_Label.csv", index=0)
```
