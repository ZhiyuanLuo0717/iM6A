### Get sequence

```python
from Bio import SeqIO
import numpy as np
import pandas as pd
```

```python
meta=[]
sequence=[]
```

```python
seq = ('FC.txt')
```

```python
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Data = pd.read_csv("FC.csv")
```

```python
Result = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Result
```

```python
Result.drop(["Meta"], axis=1, inplace=True)
```

```python
Result["Length"] = Result["txEnd"] - Result["txStart"]
```

```python
Result = Result.sort_values(by=["chrom","txStart"], axis=0, ascending=True)
Result = Result.reset_index(drop = True)
```

```python
Result
```

```python
Positive = Result[Result["strand"]=="+"]
Positive = Positive.reset_index(drop = True)
Negative = Result[Result["strand"]=="-"]
Negative = Negative.reset_index(drop = True)
```

```python
Positive
```

```python
Positive["Start"] = Positive["cdsStart"] - Positive["txStart"]
Positive["Stop"] = Positive["cdsEnd"] - Positive["txStart"]
```

```python
Sequence = Positive["Sequence"].tolist()
Start = Positive["Start"].tolist()
Stop = Positive["Stop"].tolist()
```

```python
New = []

for i in range(len(Sequence)):
    input_sequence = Sequence[i]
    seq = input_sequence[(Stop[i]-3):(Stop[i])]
    seq = seq.upper()
    New.append(seq)

New = np.array(New)
New = pd.DataFrame({"StopCodon":New})
```

```python
Positive["StopCodon"] = 0
Positive["StopCodon"] = New["StopCodon"]
```

```python
Positive
```

```python
Negative["Start"] = Negative["txEnd"] - Negative["cdsEnd"]
Negative["Stop"] = Negative["txEnd"] - Negative["cdsStart"]
```

```python
Sequence = Negative["Sequence"].tolist()
Start = Negative["Start"].tolist()
Stop = Negative["Stop"].tolist()
```

```python
from Bio.Seq import Seq
```

```python
New = []
for i in range(len(Sequence)):
    input_sequence = Sequence[i][::-1]
    seq = Seq(input_sequence).complement()
    seq = seq[(Stop[i]-3):(Stop[i])]
    seq = str(seq)
    seq = seq.upper()
    New.append(seq)

New = np.array(New)
New = pd.DataFrame({"StopCodon":New})
```

```python
Negative["StopCodon"] = 0
Negative["StopCodon"] = New["StopCodon"]
```

```python
Negative
```

```python
Data = Positive.append(Negative)
Data = Data.reset_index(drop = True)
```
