## Grab sequence

```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_csv("UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority.txt", sep="\t", header=None)
```

```python
Data
```

```python
Data.drop([0,11,13,14,15], axis=1, inplace=True)
```

```python
Data
```

```python
Data[1] = Data[12]
Data.drop([12], axis=1, inplace=True)
```

```python
Data
```

```python
Data.columns = ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]
```

```python
Data
```

### Select coding genes

```python
idx = (Data["cdsStart"] != Data["cdsEnd"])
```

```python
Data = Data.loc[idx, ["name", "chrom", "strand", "txStart", "txEnd", "cdsStart", "cdsEnd", "exonCount", "exonStarts", "exonEnds"]]
```

```python
Data
```

### Keep standard chromosomes

```python
Data = Data[:22357]
```

```python
Data
```

```python
Data.drop(["exonCount", "exonStarts", "exonEnds"], axis=1, inplace=True)
```

```python
Data
```

```python
StopCodon = pd.read_csv("StopCodon.csv")
```

```python
StopCodon = StopCodon[["name","chrom","strand","StopCodon"]]
```

```python
Data = pd.merge(Data, StopCodon, how="inner", on=["name","chrom","strand"])
```

```python
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Group = pd.read_csv("Group.csv")
Group = Group[["name","chrom","strand","Group"]]
```

```python
Group
```

```python
Data = pd.merge(Data, Group, how="inner", on=["name","chrom","strand"])
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.to_csv("UniqmRNA_GENCODE_VM7com_mm10_RefSeqpriority_Stop_Codon.csv", index=0)
```

### Create Bed

```python
Bed = Data.iloc[:, [1, 3, 4]]
```

```python
Bed.to_csv("StopCodon.bed", sep="\t", header=False, index=False)
```
