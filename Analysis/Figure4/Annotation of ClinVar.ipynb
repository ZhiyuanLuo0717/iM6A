## ClinVar

```python
import numpy as np
import pandas as pd
```

```python
Data = pd.read_table("variant_summary.txt", low_memory=False)
```

```python
Data.columns
```

```python
Data = Data[Data["Type"]=="single nucleotide variant"]
Data = Data.reset_index(drop = True)
```

```python
Data = Data[Data["Assembly"] == "GRCh37"]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.rename(columns={"RS# (dbSNP)":"SNPID","GeneSymbol":"GENE"}, inplace = True)
```

```python
Data["SNPID"] = Data["SNPID"].astype(str)
Data["SNPID"] = "rs" + Data["SNPID"]
```

```python
SNP = pd.read_csv("ClinVar_Mark.csv")
```

```python
SNP
```

```python
Data = pd.merge(Data, SNP, how="inner", on=["SNPID","GENE"])
```

```python
Data
```

```python
Data = Data.drop_duplicates(subset=["chrom","POS","GENE","strand","SNPID"], keep="first")
Data = Data.reset_index(drop = True)
```

```python
Data = Data[["chrom","GENE","strand","SNPID","POS","Mark","ClinicalSignificance","ClinSigSimple"]]
```
