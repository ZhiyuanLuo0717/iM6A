```python
import glob,os
import numpy as np
import pandas as pd
```

### Read multiple files

```python
path = "m6AQTL"
```

```python
file = glob.glob(os.path.join(path, "*.bed"))
```

```python
d = []
for f in file:
    d.append(pd.read_table(f))
```

### Delete empty dataframe

```python
for i in range(len(d))[::-1]:
    if len(d[i]) == 0:
        del d[i]
```

```python
Data = pd.concat(d)
```

```python
# Data = Data.sort_values(by=["chrom","Start"], axis=0, ascending=True)
# Data = Data.reset_index(drop = True)
```

```python
Data
```

### Get iM6A sites

#### For positive strand

```python
idx = (Data["strand"] == "+")
```

```python
Positive = Data.loc[idx, ["name", "chrom", "strand", "Start", "End", "Probabilty"]]
```

```python
Positive["Start"] = Positive["Start"] + 1
Positive["End"] = Positive["End"] + 1
```

```python
Positive
```

#### For negative strand

```python
idx = (Data["strand"] == "-")
```

```python
Negative = Data.loc[idx, ["name", "chrom", "strand", "Start", "End", "Probabilty"]]
```

```python
Negative["Start"] = Negative["Start"] - 1
Negative["End"] = Negative["End"] - 1
```

```python
Negative
```

```python
iM6A = pd.concat([Positive, Negative], axis=0)
```

```python
iM6A = iM6A.sort_values(by=["chrom","Start"], axis=0, ascending=True)
iM6A = iM6A.reset_index(drop = True)
```

```python
iM6A
```

```python
iM6A = iM6A[["chrom", "Start", "End", "name", "Probabilty", "strand"]]
```

```python
iM6A["End"] = iM6A["End"] + 1
```

```python
iM6A
```

```python
iM6A["Start"] = iM6A["Start"].astype(int)
iM6A["End"] = iM6A["End"].astype(int)
```

```python
iM6A
```

```python
Data = pd.read_csv("iM6A_humanWhistleRAC_Label.csv")
```

```python
Data.rename(columns={"Probabilty":"ProbREF"}, inplace = True)
```

```python
Data = pd.merge(Data, iM6A, how="inner", on=["chrom","Start","End","strand"])
```
