## ClinVar

```python
Data = pd.read_table("variant_summary.txt", low_memory=False)
```

```python
Data.columns
```

```python
Data = Data[Data["Type"]=="single nucleotide variant"]
Data = Data.reset_index(drop = True)
```

```python
Data = Data[["Chromosome","RS# (dbSNP)","Start","Stop","GeneSymbol","PositionVCF","ReferenceAlleleVCF","AlternateAlleleVCF","Assembly"]]
```

```python
Data
```

```python
Data = Data[Data["Assembly"] == "GRCh37"]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data = Data[["Chromosome","RS# (dbSNP)","Start","Stop","GeneSymbol","PositionVCF","ReferenceAlleleVCF","AlternateAlleleVCF"]]
Data.columns = ["chrom","SNPID","Start","End","GENE","POS","REF","ALT"]
```

```python
Data = Data[(Data["chrom"] !="na") | (Data["SNPID"] != -1) | (Data["POS"] != -1) | (Data["REF"] != "na") | (Data["ALT"] != "na")]
Data = Data.reset_index(drop = True)
```

```python
Data = Data[Data["SNPID"] > 0]
Data = Data[Data["Start"] > 0]
Data = Data[Data["End"] > 0]
Data = Data[Data["POS"] > 0]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["SNPID"] = Data["SNPID"].astype(str)
Data["chrom"] = Data["chrom"].astype(str)
Data["SNPID"] = "rs" + Data["SNPID"]
Data["chrom"] = "chr" + Data["chrom"] 
```

```python
Data
```

```python
Data["Start"] = Data["End"] - 1
```

```python
Data = Data[Data["chrom"] != "chrMT"]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Bed = Data.iloc[:, [0, 2, 3]]
Bed.to_csv("ClinVar.bed", sep="\t", header=False, index=False)
!bedtools getfasta -bed ClinVar.bed -fi /pod/2/ke-lab/LUOZ/Reference/Fasta/hg19/hg19.fa -fo ClinVar.txt
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
meta=[]
sequence=[]
```

```python
seq = ('ClinVar.txt')
for seq_record in SeqIO.parse(seq, "fasta"):
    meta.append(str(seq_record.id))
    sequence.append(str(seq_record.seq))
```

```python
Fasta = pd.DataFrame(data ={'Meta':meta,'Sequence':sequence})
```

```python
Fasta
```

```python
Data = pd.merge(Data, Fasta, left_index=True, right_index=True, how='outer')
```

```python
Data
```

```python
Data = Data[Data["REF"] == Data["Sequence"]]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data.drop(["Meta","Sequence"], axis=1, inplace=True)
```

```python
Data
```

```python
Data = Data[["chrom", "SNPID", "GENE", "POS", "REF", "ALT"]]
```

```python
Data
```

```python
Data.to_csv("ClinVar_SNP.csv",index=0)
```

```python
Data.rename(columns={"GENE":"name"}, inplace = True)
```

```python
Data
```

```python
StopCodon = pd.read_csv("HumanStopCodon.csv")
```

```python
Data = pd.merge(Data, StopCodon, how="inner", on=["name","chrom"])
```

```python
Data
```

```python
Data = Data[(Data["POS"]>=Data["txStart"]) & (Data["POS"]<=Data["txEnd"])]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data = Data.drop_duplicates(subset=["chrom","POS","name","strand","SNPID"], keep="first")
Data = Data.reset_index(drop = True)
```

```python
Data
```
