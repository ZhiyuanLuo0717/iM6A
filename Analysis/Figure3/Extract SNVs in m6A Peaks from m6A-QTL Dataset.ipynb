### m6A Peaks from m6A-QTL

````python
m6A = pd.read_csv("m6APeak_logOR_GC.IP.adjusted_qqnorm.fastQTL.csv")
````

````python
m6A.drop(["chr","start","end"], axis=1, inplace=True)
````

````python
m6A
````

````python
for i in range(len(m6A)):
    m6A.loc[i,"chrom"] = m6A.loc[i, "PEAK"].split(":")[0]
    m6A.loc[i,"PeakStart"] = m6A.loc[i, "PEAK"].split(":")[1].split("-")[0]
    m6A.loc[i,"PeakEnd"] = m6A.loc[i, "PEAK"].split(":")[1].split("-")[1].split("_")[0]
    m6A.loc[i,"name"] = m6A.loc[i, "PEAK"].split(":")[1].split("-")[1].split("_")[1]
    m6A.loc[i,"strand"] = m6A.loc[i, "PEAK"].split("_")[-1]
````

````python
m6A["PeakStart"] = m6A["PeakStart"].astype(int)
m6A["PeakEnd"] = m6A["PeakEnd"].astype(int)
````

````python
m6A
````

````python
m6A.rename(columns={"name":"GENE","PEAK":"Peak"}, inplace = True)
````

````python
Peak = m6A[["chrom","PeakStart","PeakEnd","GENE","Peak","strand"]]
````

````python
Peak
````

````python
Peak.to_csv("m6APeak_QTL.bed", sep="\t", header=False, index=False)
````
