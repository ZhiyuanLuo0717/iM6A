### Load SNP info

```python
SNP = pd.read_csv("Fasta_m6A_QTL_SigSite.csv")
```

```python
SNP = SNP[["chrom","name","strand","SNPID","POS"]]
```

```python
SNP
```

```python
Data = pd.merge(Data, SNP, how="inner", on=["chrom","name","strand","SNPID"])
```

```python
Data["Distance"] = Data["Start"] - Data["POS"] + 1
Positive = Data[Data["strand"]=="+"]
Negative = Data[Data["strand"]=="-"]
Negative["Distance"] = -(Negative["Distance"])
Data= pd.concat([Positive, Negative], axis=0)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Data["Dvalue"] = Data["ProbALT"] - Data["ProbREF"]
Data = Data[abs(Data["Dvalue"]) >=0.1]
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
Regression = pd.read_csv("LinearRegression.csv")
```

```python
Regression
```

```python
Data = pd.merge(Data, Regression, how="inner", on=["SNPID"])
```

```python
Data
```

```python
Positive = Data[Data["Dvalue"] > 0]
Positive = Positive.reset_index(drop = True)
Positive = Positive.sort_values(by=["Dvalue"], axis=0, ascending=False)
Positive.drop_duplicates(["SNPID"], inplace=True, keep="first")
Positive = Positive.reset_index(drop = True)

Negative = Data[Data["Dvalue"] < 0]
Negative = Negative.reset_index(drop = True)
Negative = Negative.sort_values(by=["Dvalue"], axis=0, ascending=True)
Negative.drop_duplicates(["SNPID"], inplace=True, keep="first")
Negative = Negative.reset_index(drop = True)

Data = pd.concat([Positive, Negative], axis=0)
Data = Data.reset_index(drop = True)
```

```python
Data
```

```python
m6A = pd.read_csv("m6APeak_logOR_GC.IP.adjusted_qqnorm.fastQTL.csv")
m6A.drop(["chr","start","end"], axis=1, inplace=True)
m6A
```

```python
SNP = pd.read_table("SNP.txt", sep="\t", low_memory=False)
SNP.rename(columns={"#CHROM":"chrom", "ID":"SNPID"}, inplace = True)
SNP.drop(["chrom","POS","QUAL","FILTER","INFO","FORMAT"], axis=1, inplace=True)
SNP
```

```python
SNPID = Data["SNPID"].tolist()
PEAK = Data["Peak"].tolist()

df = pd.DataFrame({"Type":["0|0","1|0","1|1"]})
for i in range(len(Data)):
    Peak = m6A[m6A["PEAK"]== PEAK[i]]
    Peak = Peak.drop_duplicates()
    Peak = Peak.reset_index(drop = True)
    Peak = Peak.iloc[:,1:61]
    
    ID = SNP[SNP["SNPID"]== SNPID[i]]
    ID = ID.iloc[:,3:63]
    
    ID = ID.T
    Peak = Peak.T
    Merge = pd.merge(ID, Peak, left_index=True, right_index=True)
    Merge.columns=["Type","Value"] 
    Merge.replace("0|1", "1|0", inplace=True)
    Mean = pd.DataFrame(Merge.groupby("Type")["Value"].mean())
    Mean.rename(columns={"Value":SNPID[i]}, inplace = True)    
    df = pd.merge(df, Mean, how='left', on='Type')

df = df.T
header = df.iloc[0]
df = df[1:]
df.columns = header
df
```

```python
df.dropna(axis=0, how="any", inplace=True)
df = df.reset_index(drop = False)
df.rename(columns={"index":"SNPID"}, inplace = True)
df
```

```python
Data = pd.merge(Data, df, how="inner", on=["SNPID"])
```
